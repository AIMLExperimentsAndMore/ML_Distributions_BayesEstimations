```python
import numpy as np
import matplotlib.pyplot as plt
```

# Drawing the Gauss Distribution based on analytical function

```python
# Define the Gaussian function
def gaussian(x, mean, sigma):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / sigma) ** 2)
```

```python
# Parameters for the Gaussian
mean = 0  # Mean of the Gaussian
sigma = 1  # Standard deviation (sigma)

# Generate x values
x = np.linspace(mean - 4 * sigma, mean + 4 * sigma, 500)
y = gaussian(x, mean, sigma)
```

```python
# Create the plot
plt.figure(figsize=(10, 6))
# Plot the Gaussian curve
plt.plot(x, y, label='Gaussian Curve', color='blue')
# Add a vertical line for the mean
plt.axvline(mean, color='red', linestyle='-', linewidth=2, label='Mean (µ)')
# Add dashed vertical lines for 1-sigma boundaries
plt.axvline(mean - sigma, color='green', linestyle='--', linewidth=1.5, label='-1σ')
plt.axvline(mean + sigma, color='green', linestyle='--', linewidth=1.5, label='+1σ')

# Customize the plot
plt.xlabel('x', fontsize=12)
plt.ylabel(r'$N(x|\mu, \sigma)$', fontsize=12)
plt.legend(fontsize=10)
plt.grid(alpha=0.4)

# Show the plot
plt.show()
```

# Drawing the Gauss Distribution based on generated Random Data

```python
# Generate Gaussian white noise
num_samples = 10000
mean = 0  # Mean of the Gaussian distribution
sigma = 1  # Standard deviation
white_noise = np.random.normal(mean, sigma, num_samples)

# Creating a Histogram
# Define amplitude bins from -2 to 2 with 0.1 intervals
bins = np.arange(-2, 2.1, 0.1)  # Bins from -2 to 2 in steps of 0.1
bin_counts = np.zeros(len(bins) - 1)

# Count the number of samples in each bin
for i in range(len(bins) - 1):
    bin_counts[i] = np.sum((white_noise >= bins[i]) & (white_noise < bins[i + 1]))

# Normalize bin counts so the total sum is 1
bin_counts /= np.sum(bin_counts)

# Plot the Gaussian distribution as beams
plt.figure(figsize=(10, 6))
for i in range(len(bin_counts)):
    bin_center = (bins[i] + bins[i + 1]) / 2
    beam_height = bin_counts[i]
    plt.bar(bin_center, beam_height, width=0.08, color="blue", alpha=0.7)

# Add Gaussian curve for reference
x = np.linspace(-3, 3, 500)
y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / sigma) ** 2)
plt.plot(x, y * (bins[1] - bins[0]), label='Gaussian Curve', color='red', linewidth=2)

# Customize the plot
plt.title('1D Gaussian Plot with Normalized Binned Amplitudes', fontsize=14)
plt.xlabel('Amplitude', fontsize=12)
plt.ylabel('Normalized Beam Height', fontsize=12)
plt.legend(fontsize=10)
plt.grid(alpha=0.4)

# Show the plot
plt.show()
```
